## sub-matrix: Safety

Safety is one of the most important part of value to assess whether the neighborhood is a good one. We make a graph of the counts of arrest to show which neighborhood is the best one.

```python
import pandas as pd
import matplotlib.pyplot as plt
csv_file_path = 'resource/Crimerate/ArrestRate.csv'
data = pd.read_csv(csv_file_path)
neighborhood_arrest_counts = data['INCIDENTNEIGHBORHOOD'].value_counts()
plt.figure(figsize=(16, 8)) 
neighborhood_arrest_counts.plot(kind='bar', color='skyblue')
plt.title('Arrest Counts by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Arrest Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

In our dataset graph, we devided the graph into many categories, Data shows that a lots of popular neighborhoods like squirrel hill and shady side is having a good result of arrest rate.

## submatrix-convenience

Convenience is also an important stantard to value whether the neighborhood is good or not. So we choose a dataset that contain the count of store of each neighborhood


```python
import pandas as pd
import matplotlib.pyplot as plt
data_path = 'resource/convenience/stores.csv'  
stores_data = pd.read_csv(data_path)
store_counts = stores_data['municipality'].value_counts()
plt.figure(figsize=(19, 8))
store_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Stores by Municipality', fontsize=14)
plt.xlabel('Municipality', fontsize=9)
plt.ylabel('Number of Stores', fontsize=12)
plt.xticks(rotation=90, ha='center', fontsize=9)  
plt.tight_layout()
plt.show()
```

The result of out dataset show that ROSS TWP contains the largest amount of stores, also, some famous neighborhood like squirrel hill and ockland also contains a lot of stoures.
